```python
import matplotlib.pylab as plt        # Matplotlib for plotting
import seaborn as sns                # Seaborn for data visualization
import pandas as pd                  # Pandas for data manipulation and analysis
from sklearn.linear_model import LinearRegression  # Import Linear Regression model
from sklearn.model_selection import train_test_split  # Splitting data into train and test sets
from sklearn.metrics import r2_score  # Import R-squared metric for model evaluation
```

```python
data=pd.read_csv('/content/Private_data.csv')
data.head()
```

```python
# dropping names and Unnamed from datasets
df=data.drop(['NAME OF THE STUDENT','Unnamed: 0'],axis=1)
```

```python
#setting all columns to lower case
df.columns=df.columns.str.lower()
```

```python
df.describe()
```

```python
data.isnull().sum()
```

```python
x=df[["general management score (out of 50)", "domain specific score (out 50)", "total score (out of 100)"]]
y=df[["rank", "percentile"]]
x,y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
m_pred=lr.predict(x_test)
```

```python
r2 = r2_score(y_test, m_pred)
print(f"R-squared score: {r2}")
```

```python
n = len(y_test) # Number of observations
p = x_test.shape[1] # Number of features
adjusted_r2 = 1 - (1-r2)*(n-1)/(n-p-1)
print(f"Adjusted R-squared score: {adjusted_r2}")
```
